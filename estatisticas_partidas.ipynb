```python
import chess.pgn
import os

pgn = open('partidas/'+os.listdir('partidas')[0], encoding='utf-8')
offsets = []
partidas = []

while True:
    offset = pgn.tell()

    headers = chess.pgn.read_headers(pgn)
    if headers is None:
        break

    offsets.append(offset)
    
for offset in offsets:
    pgn.seek(offset)
    partidas.append(chess.pgn.read_game(pgn))
```

```python
import pandas as pd
headers = pd.DataFrame([partida.headers for partida in partidas])
headers.head()
```

```python
board = chess.Board()
for move in partidas[-1].mainline_moves():
    board.push(move)
    #display(board)
```
